testing new patient

```python
# 1. Import libraries
import pandas as pd
import joblib

# 2. Load your saved best model
best_model = joblib.load("best_model.pkl")

# 3. Create a new patient data example
# Replace values with actual patient info
new_patient = pd.DataFrame({
    'age': [55],
    'gender': ['male'],
    'blood_pressure': [140],
    'specific_gravity': [1.020],
    'albumin': [1],
    'sugar': [0],
    'pus_cell': ['normal'],
    'pus_cell_clumps': ['absent'],
    'bacteria': ['absent'],
    'blood_glucose_random': [150],
    'blood_urea': [30],
    'serum_creatinine': [1.2],
    'sodium': [140],
    'potassium': [4.5],
    'hemoglobin': [13.5],
    'packed_cell_volume': [40],
    'white_blood_cell_count': [8000],
    'red_blood_cell_count': [4.8],
    'hypertension': ['yes'],
    'diabetes_mellitus': ['no'],
    'coronary_artery_disease': ['no'],
    'appetite': ['good'],
    'anemia': ['no'],
    'pedal_edema': ['no']
})

# 4. Encode categorical variables if needed
# Example for binary encoding
new_patient['gender'] = new_patient['gender'].map({'male':1, 'female':0})
new_patient['hypertension'] = new_patient['hypertension'].map({'yes':1, 'no':0})
new_patient['diabetes_mellitus'] = new_patient['diabetes_mellitus'].map({'yes':1, 'no':0})
new_patient['coronary_artery_disease'] = new_patient['coronary_artery_disease'].map({'yes':1, 'no':0})
new_patient['appetite'] = new_patient['appetite'].map({'good':1, 'poor':0})
new_patient['anemia'] = new_patient['anemia'].map({'yes':1, 'no':0})
new_patient['pedal_edema'] = new_patient['pedal_edema'].map({'yes':1, 'no':0})
new_patient['pus_cell'] = new_patient['pus_cell'].map({'normal':0, 'abnormal':1})
new_patient['pus_cell_clumps'] = new_patient['pus_cell_clumps'].map({'absent':0, 'present':1})
new_patient['bacteria'] = new_patient['bacteria'].map({'absent':0, 'present':1})

# 5. Make prediction
prediction = best_model.predict(new_patient)
prediction_proba = best_model.predict_proba(new_patient)  # gives probability if model supports it

# 6. Show result
if prediction[0] == 1:
    print("⚠️ The patient is predicted to have CKD.")
else:
    print("✅ The patient is predicted NOT to have CKD.")

print("Prediction probability:", prediction_proba)
```
